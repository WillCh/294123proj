```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# prepare the data
p = 100
n = 4000
n_valid = 200
mu = 0
sigma = 1

x1 = np.random.rand(p,1)
x2 = np.random.rand(p,1)

prob = np.random.rand(n,1)
A = np.random.rand(n, p)
y = np.zeros((n,1))

error = np.random.normal(mu, sigma, n)
A_v = np.random.rand(n_valid, p)
y_v = np.zeros((n_valid, 1))
p_v = np.random.rand(n_valid, 1)
e_v = np.random.normal(mu, sigma, n_valid)
for i in xrange(n):
    if prob[i] > 0.5:
        y[i,0] = np.sum(A[i,:]*x1) + error[i]
    else:
        y[i,0] = np.sum(A[i,:]*x2) + error[i]

for i in xrange(n_valid):
    if prob[i] > 0.5:
        y_v[i,0] = np.sum(A_v[i,:]*x1) + e_v[i]
    else:
        y_v[i,0] = np.sum(A_v[i,:]*x2) + e_v[i]
data_v = (A_v, y_v)
```

```python
# try the partition of data to the nodes
nodeN = 4
# use the hash table to hold the partitioned data
data = {}
for i in xrange(nodeN):
    s = n/nodeN*i
    e = n/nodeN*(i+1)
    datai = A[s:e, :], y[s:e, 0]
    data[i] = datai
```

```python
# define the cost and gradient function here
def cost(x, data):
    """
        data is a tuple with 0: A, and 1: y
        L2 norm here, dim of x is p*1
    """
    A = data[0]
    y = data[1]
    n, p = A.shape
    ans = sum((np.dot(A,x).reshape(n, 1) - y.reshape(n,1))**2)
    assert len(ans) == 1
    return ans
def gradient(x, data):
    """
        return a vector which has dim p *1
        the gradient is 2A.t(Ax-y)
    """
    A = data[0]
    y = data[1]
    n, p = A.shape
    assert n == len(y)
    g = 2*np.dot(np.transpose(A), (np.dot(A,x) - y.reshape(n,1)))
    g_1, g_2 = g.shape
    assert g_1 == p and g_2 == 1
    return g
    
def one_step_grad(x, data, stepSize):
    """
        do one step gradient descent
    """
    g = gradient(x, data)
    return x - stepSize*g
```

```python
# define the simu-distributed method

def simu_distri(data, data_v, nodeN, x0, combine, generate, maxIter, stepSize = 0.00001):
    """
        output the cost at each iteration,
    """
    # init
    x_1 = {}
    x_2 = {}
    for i in xrange(nodeN):
        x_1[i] = x0
    # start iteration
    costRes = np.zeros(maxIter)
    for iter in xrange(maxIter):
        for i in xrange(nodeN):
            x_2[i] = one_step_grad(x_1[i], data[i], stepSize*1.0/(iter+1))
        # reduce the x
        x_1 = combine(x_2, data, nodeN, iter)
        tmpCost = 0.0
        x_pred = generate(data_v, x_1, nodeN, iter)
        costRes[iter] = cost(x_pred, data_v)
    return costRes
```
